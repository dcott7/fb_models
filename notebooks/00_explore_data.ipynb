```python
from fb_models.dataset import (
    load_pbp_dataset, 
    load_participation_dataset, 
    load_players_dataset, 
    load_weekly_rosters_dataset,
    load_games_dataset,
)
from fb_models.features.feature_store import FeatureStore
```

```python
# Seasons 
SEASONS = list(range(2019, 2026))
```

```python
pbp = load_pbp_dataset(seasons=SEASONS)
participation = load_participation_dataset(seasons=SEASONS)
players = load_players_dataset()
weekly_rosters = load_weekly_rosters_dataset(seasons=SEASONS)
games = load_games_dataset()
```

```python
fs = FeatureStore(
    plays_df=pbp,
    participation_df=participation,
    players_df=players, 
    weekly_rosters_df=weekly_rosters,
    games_df=games,
)
```

```python
# build all of the feature data
fs.build()
```
